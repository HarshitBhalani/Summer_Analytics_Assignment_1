### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
import pandas as pd
df = pd.read_csv('Cars.csv')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
median_mpg = df['mpg'].median()
df_above_median = df[df['mpg'] > median_mpg].copy()
df_above_median['hp_wt_ratio'] = df_above_median['horsepower'] / df_above_median['weight']
max_ratio = df_above_median['hp_wt_ratio'].max()
print(df_above_median[df_above_median['hp_wt_ratio'] == max_ratio][['name', 'mpg', 'horsepower', 'weight', 'hp_wt_ratio']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Cars.csv')

# Group by 'model_year' and 'origin' and calculate the mean MPG
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()


plt.figure(figsize=(12, 6))  
for origin in avg_mpg.columns:
    plt.plot(avg_mpg.index, avg_mpg[origin], label=origin)
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.title('Average MPG by Model Year and Origin')
plt.legend(title='Origin')
plt.grid(True)
plt.xticks(avg_mpg.index)  
plt.tight_layout() 
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Cars.csv')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

color_map = {'japan': 'red', 'europe': 'green', 'usa': 'blue'}
df['color'] = df['origin'].map(color_map)

plt.figure(figsize=(10,6))
for origin in ['japan', 'europe', 'usa']:
    subset = df[df['origin'] == origin]
    plt.scatter(
        subset['horsepower'],
        subset['weight'],
        s=subset['mpg'] * 5, 
        c=color_map[origin],
        alpha=0.7,
        label=origin
    )

plt.title('Horsepower vs Weight by Origin (Point Size = MPG)')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd
df = pd.read_csv('Cars.csv')

grouped = df.groupby(['name', 'model_year'])['mpg'].mean().reset_index()

name_counts = grouped['name'].value_counts()
multi_year_names = name_counts[name_counts > 1].index

results = []
for name in multi_year_names:
    mpgs = grouped[grouped['name'] == name]['mpg']
    std = mpgs.std()
    if std < 1.0:
        appearances = (df['name'] == name).sum()
        avg_mpg = df[df['name'] == name]['mpg'].mean()
        results.append({
            'name': name,
            'appearances': appearances,
            'avg_mpg': avg_mpg,
            'std_mpg': std
        })
if results:
    results_df = pd.DataFrame(results)
    min_std = results_df['std_mpg'].min()
    final = results_df[results_df['std_mpg'] == min_std]
    # Bonus: sort
    final = final.sort_values(['appearances', 'avg_mpg'], ascending=[False, False])
    # Round for neatness
    final['avg_mpg'] = final['avg_mpg'].round(2)
    final['std_mpg'] = final['std_mpg'].round(2)
    print(final[['name', 'appearances', 'avg_mpg']])
else:
    print("No consistent car models found.")
```
